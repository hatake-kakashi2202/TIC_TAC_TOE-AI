#### Pygame Functions

```python
# returns the co-ordinate of x,y cell in which mouse is clicked

def map_mouse_to_board(x, y):
    #column retrival
    for i in range(N):
        if margin + (gameSize / N) * i <= x < margin + (gameSize / N) * (i + 1):
            column = i
    #row retrival
    for i in range(N):    
        if margin + (gameSize / N) * i <= y < margin + (gameSize / N) * (i + 1):
            row = i
    return row,column
            
#function for drawing lines on board(to make tictactoe playyable)   
    
def draw_lines():
    # vertical lines
    for i in range(0,N+1):
        pygame.draw.line(screen, lineColor, (margin + (gameSize // N) * i, margin),
                         (margin + (gameSize // N) * i, screenSize - margin), lineSize)
    # horizontal lines
        pygame.draw.line(screen, lineColor, (margin, margin + (gameSize // N) * i),
                         (screenSize - margin, margin + (gameSize // N) * i), lineSize)

# function for drawing board as program board contains -1,1,0 in cells 
# to render them as X,O on to display it is called        

def draw_board(state):
    
    chars = {
        -1: h_choice,
        1: c_choice,
        0: ' '
    } 
    myFont = pygame.font.SysFont('Tahoma', gameSize // N)
    x = 0
    for row in state:
        y = 0
        for cell in row:
            if cell == h_choice:
                cell = -1
            elif cell == c_choice:
                cell = +1

            symbol = chars[cell]
            sentstring = ''
            color = ''
            
            if symbol == xMark:
                color = xColor
                sentstring = 'X'
            elif symbol == oMark:
                color = oColor
                sentstring = 'O'
            else:
                color = oColor
                sentstring = ''
            
            text_surface = myFont.render(sentstring, False, color)
            screen.blit(text_surface, (y * (gameSize //N) + margin + (gameSize // (N * 6)), x * (gameSize // N) + margin/2))
            pygame.display.update() 
            y = y + 1
        pygame.display.update() 
        x = x + 1
```

#### Main game Functions

```python
# General evaluation function that returns 1,0,-1

def evaluate(boardstate):
    if goal_test(boardstate, COMP):
        score = +1
    elif goal_test(boardstate, HUMAN):
        score = -1
    else:
        score = 0

    return score

# checks for the elements of my_dicy_comp and my_dict_human and assigns the score from dictionary

def heauristic(boardstate):
    win_state = [
        [boardstate[0][0], boardstate[0][1], boardstate[0][2]],
        [boardstate[1][0], boardstate[1][1], boardstate[1][2]],
        [boardstate[2][0], boardstate[2][1], boardstate[2][2]],
        [boardstate[0][0], boardstate[1][0], boardstate[2][0]],
        [boardstate[0][1], boardstate[1][1], boardstate[2][1]],
        [boardstate[0][2], boardstate[1][2], boardstate[2][2]],
        [boardstate[0][0], boardstate[1][1], boardstate[2][2]],
        [boardstate[2][0], boardstate[1][1], boardstate[0][2]],
    ]
    open_paths_comp = 0
    open_paths_human = 0
    my_dict_comp = {1:[[1,0,0],[0,1,0],[0,0,1]],10:[[1,1,0],[1,0,1],[0,1,1]],100:[[1,1,1]]}
    my_dict_human = {1:[[-1,0,0],[0,-1,0],[0,0,-1]],10:[[-1,-1,0],[-1,0,-1],[0,-1,-1]],100:[[-1,-1,-1]]}
    for i in win_state:
        for j in my_dict_comp:
            if i in my_dict_comp[j]:
                open_paths_comp+=j
        for j in my_dict_human:
            if i in my_dict_human[j]:
                open_paths_human-=j
    score = open_paths_comp+open_paths_human
    return score

# Simple goal test function

def goal_test(boardstate, player):
    win_state = [
        [boardstate[0][0], boardstate[0][1], boardstate[0][2]],
        [boardstate[1][0], boardstate[1][1], boardstate[1][2]],
        [boardstate[2][0], boardstate[2][1], boardstate[2][2]],
        [boardstate[0][0], boardstate[1][0], boardstate[2][0]],
        [boardstate[0][1], boardstate[1][1], boardstate[2][1]],
        [boardstate[0][2], boardstate[1][2], boardstate[2][2]],
        [boardstate[0][0], boardstate[1][1], boardstate[2][2]],
        [boardstate[2][0], boardstate[1][1], boardstate[0][2]],
    ]
    if [player, player, player] in win_state:
        return True
    else:
        return False
    
# returns true if move placed by player is valid
    
def game_over(boardstate):
    return goal_test(boardstate, HUMAN) or goal_test(boardstate, COMP)

def valid(x,y):
    if board[x][y] == 0:
        return True
    return False

# returns number of empty cells that are present in board(coordinates where board[x][y]==0)

def empty_cells(boardstate):
    empty = []
    for x, row in enumerate(boardstate):
        for y, cell in enumerate(row):
            if cell == 0:
                empty.append([x, y])
    return empty

#Normal minimax (both min and max are combined to one)

def minimax(boardstate, depth, player):
    if player == COMP:
        best = [-1, -1, -infinity]
    else:
        best = [-1, -1, +infinity]

    if depth == 0 or game_over(boardstate):
        score = evaluate(boardstate)
        return [-1, -1, score]
    
    for cell in empty_cells(boardstate):
        x, y = cell[0], cell[1]
        boardstate[x][y] = player
        score = minimax(boardstate, depth - 1, -player)
        boardstate[x][y] = 0
        score[0], score[1] = x, y
        # max value
        if player == COMP:
            if score[2] > best[2]:
                best = score
        # min value
        else:
            if score[2] < best[2]:
                best = score
    return best

#DepthLimit minimax

def minimax_depth(boardstate, depth, player):
    if player == COMP:
        best = [-1, -1, -infinity]
    else:
        best = [-1, -1, +infinity]

    if depth == 0 or game_over(boardstate):
        score = heauristic(boardstate)
        return [-1, -1, score]
    
    for cell in empty_cells(boardstate):
        x, y = cell[0], cell[1]
        boardstate[x][y] = player
        score = minimax_depth(boardstate, depth - 1, -player)
        boardstate[x][y] = 0
        score[0], score[1] = x, y
        # max value
        if player == COMP:
            if score[2] > best[2]:
                best = score
        # min value
        else:
            if score[2] < best[2]:
                best = score
    return best

#Minimax with Alpha Beta Pruning

def minimax_alpha_beta(boardstate, depth, player, alpha, beta):
    alpha_new = alpha
    beta_new = beta
    if player == COMP:
        best = [-1, -1, -infinity, alpha, beta]
    else:
        best = [-1, -1, +infinity, alpha, beta]

    if depth == 0 or game_over(boardstate):
        score = evaluate(boardstate)
        return [-1, -1, score, score, score]
    
    for cell in empty_cells(boardstate):
        x, y = cell[0], cell[1]
        boardstate[x][y] = player
        score = minimax_alpha_beta(boardstate, depth - 1, -player, alpha_new , beta_new )
        boardstate[x][y] = 0
        score[0], score[1] = x, y
        
        # max value
        if player == COMP:
            if score[2] > best[2]:
                best = score  
            if score[2] > alpha_new:
                alpha_new = score[2]
        # min value
        else:
            if score[2] < best[2]:
                best = score
            if score[2] < beta_new:
                beta_new = score[2]
        #alpha beta
        if alpha_new > beta_new:
            break             
    return best

#Minimax with Alpha Beta Pruning and Depth Limit

def minimax_depth_alpha_beta(boardstate, depth, player, alpha, beta):
    alpha_new = alpha
    beta_new = beta
    if player == COMP:
        best = [-1, -1, -infinity, alpha, beta]
    else:
        best = [-1, -1, +infinity, alpha, beta]
        
    if depth == 0 or game_over(boardstate):
        score = heauristic(boardstate)
        return [-1, -1, score]
    
    for cell in empty_cells(boardstate):
        x, y = cell[0], cell[1]
        boardstate[x][y] = player
        score = minimax_depth_alpha_beta(boardstate, depth - 1, -player,alpha_new,beta_new)
        boardstate[x][y] = 0
        score[0], score[1] = x, y
        
        #max value
        if player == COMP:
            if score[2] > best[2]:
                best = score  
            if score[2] > alpha_new:
                alpha_new = score[2]
        # min value
        else:
            if score[2] < best[2]:
                best = score
            if score[2] < beta_new:
                beta_new = score[2]
        #alpha beta
        if alpha_new > beta_new:
            break             
    return best

# function for Ai to make turns

def ai_turn(algo,first):
    depth = len(empty_cells(board))
    if depth == 0 or game_over(board):
        return
    if (depth == 9) and (algo != 5):
        x = choice([0,1,2])
        y = choice([0,1,2])
    else:
        if(algo == 1):
            move = minimax_alpha_beta(board, depth, COMP, -infinity,+infinity)
        elif(algo == 2):
            move = minimax(board,depth,COMP)
        elif(algo == 3):
            if(first =='N'):
                if(depth>=7 ):
                    depth = 7
            else:
                if(depth>=6):
                    depth = 6
            move = minimax_depth(board,depth,COMP)
        elif(algo == 4):
            if(first =='N'):
                if(depth>=7 ):
                    depth = 7
            else:
                if(depth>=6):
                    depth = 6
            move = minimax_depth_alpha_beta(board,depth, COMP, -infinity,+infinity)
        elif(algo == 5):
            if depth == 9:
                move = choice([[0,0],[0,2],[2,0],[2,2]])
            else:
                if(first =='N'):
                    if(depth>=7 ):
                        depth = 7
                else:
                    if(depth>=6):
                        depth = 6  
                move = minimax_depth_alpha_beta(board,depth, COMP, -infinity,+infinity)
        x, y = move[0], move[1]
    board[x][y] = COMP
    time.sleep(1)

    

def main():
    global N,board,h_choice,c_choice,first
    #original game board
    draw_lines()
    pygame.display.update()
    #game
    c_choice = ''
    canPlay = True
    # Setting computer's choice
    if h_choice == 'X':
        c_choice = 'O'
    else:
        c_choice = 'X'
    myFont = pygame.font.SysFont('Tahoma', 20)
    rect1 = pygame.Rect( margin-5,screenSize-margin-5,360, 30 )

    # Main loop of this game
    
    #game Loop
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.display.quit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_r:
                    screen.fill(backgroundColor)
                    draw_lines()
                    canPlay = True
                if event.key == pygame.K_ESCAPE:
                    pygame.display.quit()
            pygame.display.update()
            while len(empty_cells(board)) > 0 and not game_over(board):
                for event in pygame.event.get():
                    if event.type == pygame.QUIT:
                        pygame.display.quit()
                screen = pygame.display.get_surface()
                if first == 'N':
                    start_time = time.time()
                    ai_turn(algo,first)
                    block_placed.play()
                    end_time = time.time()
                    time_diff = float("{:.9f}".format(end_time - start_time-1)) 
                    taken_time = "agent taken time: "+str(time_diff)
                    pygame.draw.rect(screen,(255,255,255),rect1)
                    pygame.display.flip() 
                    text_surface = myFont.render(taken_time, True,(255,0,0))
                    screen.blit(text_surface, (margin,screenSize-margin))
                    draw_board(board)
                    pygame.display.update() 
                    first = ''

                #Human
                can_move = False
                while(not can_move):
                    for event in pygame.event.get():
                        if event.type is pygame.MOUSEBUTTONDOWN and canPlay:
                            (mouseX, mouseY) = pygame.mouse.get_pos()
                            x,y = map_mouse_to_board(mouseX, mouseY)
                            can_move = valid(x,y)
                            break  
                    if can_move:
                        break
                board[x][y] = HUMAN
                draw_board(board)
                block_placed.play()
                
                #Ai
                
                start_time = time.time()
                ai_turn(algo,first)
                block_placed.play()
                end_time = time.time()
                time_diff = float("{:.9f}".format(end_time - start_time-1)) 
                taken_time = "agent taken time: "+str(time_diff)
                pygame.draw.rect(screen,(255,255,255),rect1)
                pygame.display.flip() 
                text_surface = myFont.render(taken_time, True,(255,0,0))
                screen.blit(text_surface, (margin,screenSize-margin))
                draw_board(board)
                pygame.display.update() 
            
            # Game over message
            
            pygame.mixer.music.stop()
            if goal_test(board, HUMAN):
                myFont = pygame.font.SysFont('Tahoma', screenSize // 10)
                screen.blit(backgroundImage,[0,0])
                text_surface = myFont.render(" won!", False, (127, 140, 227))
                screen.blit(text_surface, ( screenSize // 2.5, screenSize // 2.5 ))
                pygame.display.update()
                win_sound.play()
                time.sleep(1)
                canPlay = False
                return

            elif goal_test(board, COMP):
                myFont = pygame.font.SysFont('Tahoma', screenSize // 10)
                screen.blit(backgroundImage,[0,0])
                text_surface = myFont.render("Lost!", False, (250, 164, 97))
                screen.blit(text_surface, ( screenSize // 2.5, screenSize // 2.5 ))
                pygame.display.update()
                lost_sound.play()
                time.sleep(1)
                canPlay = False
                return

            else:
                myFont = pygame.font.SysFont('Tahoma', screenSize // 10)
                screen.blit(backgroundImage,[0,0])
                text_surface = myFont.render("Draw!", False, (101, 245, 247))
                screen.blit(text_surface, (screenSize // 2.5, screenSize // 2.5 ))
                pygame.display.update()
                draw_sound.play()
                time.sleep(1.5)
                canPlay = False
                return
        pygame.display.update()
```
